# 分5組

```python
g1 = ['MLPAP_TJ-0706C01','MLPAP_TJ-0716B01','MLPAP_TJ-0716E01']
g2 = ['MLPAP_TJ-0706A02','MLPAP_TJ-0706E01','MLPAP_TJ-0716A01']
g3 = ['MLPAP_TJ-0706A01','MLPAP_TJ-0716D01']
g4 = ['MLPAP_TJ-0706D01','MLPAP_TJ-0716D02']
g5 = ['MLPAP_TJ-0706B01','MLPAP_TJ-0716C01']
```

# load df

```python
df = pd.read_csv('../data/train_cleaned.csv',index_col=0)
```

# kde

```python
for g in [g1,g2,g3,g4,g5]:
    for i in g:
        h = tag.loc[tag.TAG==i,'chinese'].values[0].split('高度')[1][1:-1]
        df[i].sample(1000).plot(kind='kde',label=h)
    plt.legend()
    plt.show()
```

# line

```python
for idx,g in enumerate([g1,g2,g3,g4,g5]):
    X = df[g].dropna(axis=0).mean(axis=1)
    plt.plot(X.values,label='g{}'.format(idx+1))
plt.legend()
plt.show()
```

```python
plt.plot((df['MLPAP_FQ-0619.PV'].shift(36)/df['MLPAP_FQI-0997.PV']).values,label='feed')
roll = (df['MLPAP_FQ-0619.PV'].shift(36)/df['MLPAP_FQI-0997.PV']).rolling(365).median().values
plt.plot(roll,label='roll')
plt.legend()
plt.show()
```

```python
table = pd.DataFrame()
table['g1'] = df[g1].mean(axis=1).values
table['g2'] = df[g2].mean(axis=1).values
table['g3'] = df[g3].mean(axis=1).values
table['g4'] = df[g4].mean(axis=1).values
table['g5'] = df[g5].mean(axis=1).values
table['target'] = (df['MLPAP_FQ-0619.PV'].shift(36)/df['MLPAP_FQI-0997.PV']).rolling(3).median().values
table = table.dropna(axis=0)
table.head()
```
